#### 1.1. Скачать данные по ссылке kc_house_data.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Считываем данные и выводим первые пять строк

houses = pd.read_csv('./kc_house_data.csv')
houses.head()
```

#### 1.2 Изучите стоимости недвижимости.

```python
plt.figure(figsize=(6, 4))
plt.hist(houses['price'], bins=30)
plt.title('Анализ стоимости недвижимости')
plt.xlabel('Стоимость, у.е.')
plt.ylabel('Количество домов')
```

Исходя из графика можно сделать вывод, что большая часть домов стоит в районе 540-550 тысяч у.е.

#### 1.3 Изучите распределение квадратуры жилой площади.

```python
plt.figure(figsize=(6, 4))
plt.hist(houses['sqft_living'], bins=30)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Жилая площадь, кв. футы')
plt.ylabel('Количество домов')
```

Таким образом можно сделать вывод, что большинство домов имеют среднюю площадь в районе 2000 кв. футов (около 180 кв. м).

#### 1.4 Изучите распределение года постройки.

```python
plt.figure(figsize=(6, 4))
plt.hist(houses['yr_built'], bins=30)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
```

Глядя на график, можно сделать следующие выводы:


1. Строительство домов из данного списка началось в 1900 году.
2. С течением лет наблюдается общая тендеция роста в строительстве домов.
3. Спад строительства наблюдался в 1930-х годах.
4. Пик строительства приходится на начало 2000-х.



## Условие 2: Задача 2

#### 2.1 Изучите распределение домов от наличия вида на набережную. Постройте график.Сделайте выводы.

```python
# Подготовим данные для создания графика

waterfront_distribution = houses['waterfront'].value_counts()
waterfront_distribution
```

```python
plt.figure(figsize=(8, 8))
plt.pie(waterfront_distribution.values, autopct='%1.1f%%')
plt.title('Распределение домов от наличия вида на набережную')
plt.legend(['Без вида', 'С видом']);
```

Из графика следует, что домов с видом на набережную менее 1%.

#### 2.2 Изучите распределение этажей домов.

```python
floors_distribution = houses['floors'].value_counts()
floors_distribution
```

```python
plt.figure(figsize=(8, 8))
plt.pie(floors_distribution.values, autopct='%1.1f%%')
plt.title('Распределение этажей домов')
plt.legend(floors_distribution.index);
```

Видно, что практически 50% составляют одноэтажные дома. Также порядка 40% приходится на двухэтажные дома. Почетное третье место заняли одноэтажные дома с мансардой.

#### 2.3 Изучите распределение состояния домов.

```python
condition_distribution = houses['condition'].value_counts()
condition_distribution
```

```python
plt.figure(figsize=(8, 8))
plt.pie(condition_distribution.values, autopct='%1.1f%%')
plt.title('Распределение состояния домов')
plt.legend(condition_distribution.index);
```

Здесь также результаты очевидны: большинство домов находятся в "хорошем" и "удовлетворительном" состоянии.

## Условие 3: Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

Возьмем в качестве ключевых параметров общую площадь, жилую площадь, этажность, состояние дома и год постройки. Выясним, как будет меняться цена от этих характеристик.

```python
plt.figure(figsize=(6, 4))
plt.scatter(houses['price'], houses['sqft_lot'])
plt.title('Зависимость цены дома от общей площади')
plt.xlabel('Цена, у.е.')
plt.ylabel('Общая площадь дома, кв. футы');
```

Из полученного графика видно, что цена не имеет линейной зависимости от общей площади дома, т.е. с ростом общей площади цена необязательно должна увеличиваться. Дома с небольшой общей площадью могут стоить довольно дорого, в то время как дома с максимальной общей площадью имеют вполне умеренную цену. Это обусловлено тем, что общая площадь не является единственным и ключевым параметром, влияющим на формирование цены.

```python
plt.figure(figsize=(6, 4))
plt.scatter(houses['price'], houses['sqft_living'])
plt.title('Зависимость цены дома от жилой площади')
plt.xlabel('Цена, у.е.')
plt.ylabel('Жилая площадь дома, кв. футы');
```

Полученные результаты показывают, что жилая площадь является более значимым параметром, чем общая площадь. Чем больше жилая площадь, тем, в общем и целом, дороже дом. Поэтому рост цены от жилой площади вполне закономерен.

```python
plt.figure(figsize=(6, 4))
plt.scatter(houses['price'], houses['floors'])
plt.title('Зависимость цены дома от кол-ва этажей')
plt.xlabel('Цена, у.е.')
plt.ylabel('Кол-во этажей');
```

Линейная зависимость цены четко прослеживается для каждой категории домов с определенной этажностью и зависит от других характеристик домов, влияющих на ценообразование. Как видно из графика, двухэтажные дома, например, могут стоить, как очень дешево, так и весьма дорого, и быть дороже трехэтажных домов. 

```python
plt.figure(figsize=(6, 4))
plt.scatter(houses['price'], houses['condition'])
plt.title('Зависимость цены дома от его состояния')
plt.xlabel('Цена, у.е.')
plt.ylabel('Состояние дома');
```

График показывает, что, чем лучше состояние дома, тем он дороже. Но, как и в предыдущем случае, линейная зависимость цены по большей мере прослеживается именно в конкретных категориях домов.

```python
plt.figure(figsize=(6, 4))
plt.scatter(houses['price'], houses['yr_built'])
plt.title('Зависимость цены дома от года постройки')
plt.xlabel('Цена, у.е.')
plt.ylabel('Год постройки')
```

Анализируя график, можно сделать вывод, что год постройки довольно слабо влияет на ценообразование.

#### Выводы.
Получается, что из выбранных мною параметров наибольшее влияние на цену дома оказывает жилая площадь. Остальные параметры играют менее существенную роль в ценообразовании. И в подтверждении данного вывода проведем корреляционный анализ.

Создадим отдельную таблицу с интересующими нас параметрами.

```python
conclusion=houses[['price', 'sqft_lot', 'sqft_living', 'yr_built', 'floors', 'condition']]
conclusion.head(10)
```

И теперь на тепловой карте посмотрим наличие линейных связей.

```python
sns.heatmap(conclusion.corr(), annot=True, linewidths=.5, cmap='coolwarm');
```

Действительно, наблюдается взаимосвязь между ценой и жилой площадью. Она равна 0,7, и ее можно классифицировать, как "заметную". Остальные же параметры показвают "слабую" линейную связь.

**Таким образом, можно утверждать, что все характеристики по-разному влияют на конечную цену дома. Однако их важность для себя в конечном счете определяет сам покупатель.**
